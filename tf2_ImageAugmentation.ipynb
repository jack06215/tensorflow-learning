```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from pathlib import Path
from tensorflow.keras import Sequential
from tensorflow.keras.layers.experimental.preprocessing import (
    RandomFlip,
    RandomRotation,
    RandomCrop,
    RandomContrast,
)
```

```python
print('Tensorflow Version: ', tf.__version__)
print('Numpy Version: ', np.__version__)
```

```python
DATASET_PATH = Path("C:\\External\\jack0\\Downloads\\CamSeq01")
NEW_SIZE = (256, 256)
```

```python
img = cv2.imread(f'{DATASET_PATH}/0016E5_07959.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, NEW_SIZE)
```

```python
data_augmentation = Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.1),
    RandomContrast(0.7),
    RandomCrop(height=200, width=200),
])
```

```python
sample = np.expand_dims(img, axis=0)
augmented_sample = data_augmentation(sample)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(augmented_sample[0])
plt.show()
```
