```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# plt.style.use('ggplot')
np.random.seed(0)
```

```python
uniform_data = np.random.rand(20, 20)
plt.figure(figsize=(10, 10))
ax = sns.heatmap(uniform_data, square=True)
```

```python
# Plot using Facegrid, separated by smoke
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="smoker", size=5, aspect=1.5)
g.map(sns.boxplot, "sex", "total_bill",
      palette='viridis', order=['Male', 'Female'])

ax1, ax2 = g.axes[0]

ax1.axhline(10, ls='--')
ax2.axhline(30, ls='--')

ax1.text(0.5, 25, "Some text")
ax2.text(0.5, 25, "Some text")

plt.show()
```

```python
df = sns.load_dataset("dots")[['align', 'firing_rate']]
df = df.sort_values(by=['align','firing_rate'], ignore_index=True)

fig = sns.catplot(x='align', y='firing_rate', kind='swarm', data=df, height=8)
# fig.set(ylim=(0.1, 8))
ax1 = fig.axes[0][0]

ax1.axhline(40, ls='--', c=(1, 0, 0))
ax1.text(1.6, 40, "firing_rate = 40")

ax1.axhline(15, ls='--', c=(1, 0, 0))
ax1.text(1.6, 15, "firing_rate = 15")
```
